# Domain of Rule 18

```python
np.random.seed(201)
ca = ECA(18, domain_18(100))
ca.evolve(100)
field = ca.spacetime
```

```python
fig = spacetime_diagram(field, cmap=plt.cm.Greys, size=20, edgecolors='grey',
       t_min=20, t_max=40, x_min=40, x_max=70, tick_size=25, ttick_spacing=2, xtick_spacing=2, linewidth='0.01')
```

# Future lightcones

```python
recon = TopologicalReconstructor(3, 3, 1, 'flc')
```

```python
recon.scan_data(domain_18_generator(100,100, 10))
```

```python
recon.plot_statistics()
```

```python
recon.reconstruct_states()
print(recon.machine.number_of_states())
```

```python
recon.transition_scan(domain_18_generator(100,100, 10))
```

```python
recon.transition_statistics()
```

```python
right_trans = recon.machine.transitions['r']
print(len(right_trans))
left_trans = recon.machine.transitions['l']
print(len(left_trans))
forward_trans = recon.machine.transitions['t']
print(len(forward_trans))
```

```python
right = right_transition_violations(90, right_trans)
print('number of right-transition violations - {}'.format(len(right)))

left = left_transition_violations(90, left_trans)
print('number of left-transition violations - {}'.format(len(left)))

forward = forward_transition_violations(90, forward_trans)
print('number of forward-transition violations - {}'.format(len(forward)))
```

```python
machine = recon.machine
machine.filter_data(field)
```

```python
flc_field = machine.state_field
```

```python
state_overlay_diagram(field, flc_field, x_min=40, x_max=70, y_min=20, y_max=40, size=20, 
                      edgecolors="black", text_colors=plt.cm.cool, tick_size=25, ytick_spacing=2, xtick_spacing=2, linewidth='0.01')
```

# Zero-Wildcard Tiling

The domain of Rule 18 above is a stict subshift of this spacetime shift space

```python
np.random.seed(25)
field = wildcard_tiling(100,100)
```

```python
fig = spacetime_diagram(field, cmap=plt.cm.Greys, size=20, edgecolors='grey',
       t_min=20, t_max=40, x_min=60, x_max=90, tick_size=25, ttick_spacing=2, xtick_spacing=2, linewidth='0.01')
```

# Future Lightcones

```python
recon = TopologicalReconstructor(3, 3, 1, 'flc')
```

```python
recon.fast_scan(wildcard_tiling_generator(500,500, 20))
```

```python
recon.plot_statistics()
```

```python
recon.reconstruct_states()
machine = recon.machine
print(machine.number_of_states())
```

```python
machine.filter_data(field)
```

```python
flc_field = machine.state_field
```

```python
labels = {1:"F", 2:"W", 3:"X"}
state_overlay_diagram(field, flc_field, y_min=20, y_max=40, x_min=60, x_max=90, size=20, 
                      edgecolors="black", text_colors=plt.cm.cool, tick_size=25, ytick_spacing=2, xtick_spacing=2, 
                      linewidth='0.01', state_labels=labels)
```
