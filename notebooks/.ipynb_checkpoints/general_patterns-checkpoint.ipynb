## V Presentation

```python
recon = TopologicalReconstructor(8,3,1, 'V_f')
```

```python
recon.scan_data(domain_18_ensemble(500,500, 10))
```

```python
recon.plot_statistics()
```

```python
recon.reconstruct_states()
```

```python
machine = recon.machine
print(machine.number_of_states())
```

```python
machine.filter_data(field)
```

```python
fig = spacetime_diagram(machine.state_field, cmap=plt.cm.twilight, size=20, edgecolors='black',
       t_min=10, t_max=30, x_max=30, tick_size=25, ttick_spacing=2, xtick_spacing=2, linewidth='0.01')
```

```python
state_overlay_diagram(field, machine.state_field, y_min=10, y_max=30, x_max=30, 
                      tick_size=25, ytick_spacing=2, xtick_spacing=2, linewidth='0.01', size=20, 
                      text_size=14, text_colors=plt.cm.twilight, edgecolors="black")
```

permute state labels to unbias state coloring 

```python
states = np.unique(machine.state_field)
```

```python
max_state = max(states)
permute = np.random.permutation(np.arange(max_state+1))
```

```python
permute_field = np.empty(np.shape(machine.state_field), dtype=machine.state_field.dtype)
```

```python
for i, state in np.ndenumerate(machine.state_field):
    permute_field[i] = permute[state]
```

```python
fig = spacetime_diagram(permute_field, cmap=plt.cm.twilight, size=20, edgecolors='black',
       t_min=10, t_max=30, x_max=30, tick_size=25, ttick_spacing=2, xtick_spacing=2, linewidth='0.01')
```

```python
state_overlay_diagram(field, permute_field, y_min=10, y_max=30, x_max=30, 
                      tick_size=25, ytick_spacing=2, xtick_spacing=2, linewidth='0.01', size=20, 
                      text_size=14, text_colors=plt.cm.twilight, edgecolors="black")
```
