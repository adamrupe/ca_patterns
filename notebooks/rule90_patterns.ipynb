```python
ca = ECA(90, random_state(100))
ca.evolve(100)
field = ca.spacetime
```

# Future lightcones

```python
recon = TopologicalReconstructor(3, 3, 1, 'flc')
```

```python
recon.scan_data(ECA_ensemble_generator(90, 500,500, 20, transient=0))
```

```python
recon.plot_statistics()
```

```python
recon.reconstruct_states()
print(recon.machine.number_of_states())
```

```python
recon.transition_scan(ECA_ensemble_generator(90, 500,500, 20, transient=0))
```

```python
recon.transition_statistics()
```

```python
right_trans = recon.machine.transitions['r']
print(len(right_trans))
left_trans = recon.machine.transitions['l']
print(len(left_trans))
forward_trans = recon.machine.transitions['t']
print(len(forward_trans))
```

```python
right = right_transition_violations(90, right_trans)
print('number of right-transition violations - {}'.format(len(right)))

left = left_transition_violations(90, left_trans)
print('number of left-transition violations - {}'.format(len(left)))

forward = forward_transition_violations(90, forward_trans)
print('number of forward-transition violations - {}'.format(len(forward)))
```

```python
machine = recon.machine
machine.filter_data(field)
```

```python
flc_field = machine.state_field
```

```python
state_overlay_diagram(field, flc_field, x_min=20, x_max=50, y_min=20, y_max=40, size=14, 
                      edgecolors="black")
```
